```python
import seaborn as sns
```

```python
cols=['Country','Last Update','Confirmed','Deaths','Recovered']
pp=sns.pairplot(corona_data[cols],height=1.8,aspect=1.8,
      plot_kws=dict(edgecolor='k',linewidth=.5),
      diag_kind='kde',diag_kws=dict(shade=True))
fig=pp.fig
fig.subplots_adjust(top=0.83)
```

```python
plt.scatter(corona_data['Confirmed'],corona_data['Deaths'],alpha=.3,edgecolor='w')
plt.ylabel('confirmation freq')
plt.xlabel('no of deaths experienced')

jp=sns.jointplot(x='Confirmed',y='Deaths',data=corona_data,kind='reg',space=0,height=5,ratio=4)
```

```python
fig=plt.figure(figsize=(8,6))
ax=fig.add_subplot(111,projection='3d')
xs=corona_data['Confirmed']
ys=corona_data['Deaths']
zs=corona_data['Recovered']
ax.scatter(xs,ys,zs,s=60,alpha=0.6,edgecolor='black')
ax.set_xlabel('Confirmed rates')
ax.set_ylabel('Death cases')
ax.set_zlabel('Recovered number')
```

```python
#corona_data on bubbleplots
plt.scatter(corona_data['Confirmed'],corona_data['Deaths'],s=corona_data['Recovered']*10,alpha=.4,edgecolor='black')
plt.xlabel('Confirmed rates')
plt.ylabel('Death cases')
plt.title('Corona on bubbles')
```
